### Inner Join by ID to identify details of vaccine administered to individuals who died as result of adverse event after vaccination, ordered by death date

```python
%%bigquery
select DATEDIED, VAX_DATE, VAX_NAME, VAX_TYPE, VAX_MANU 
from vaers_staging.Events evn
join vaers_staging.Vaccines vac on evn.VAERS_ID = vac.VAERS_ID
where evn.DIED = true
order by evn.DATEDIED desc
limit 12
```

### Inner Join by ID to identify site of vaccine injection (VAX_ROUTE) of individuals who died as result of advserse event, ordered by vaccination date

##### Legend: 
IM - Intramuscular \
SC - Subcutaneous (under skin) \
SYR - Syringe (Not specified futher) 

```python
%%bigquery
select VAX_DATE, ONSET_DATE, VAX_ROUTE 
from vaers_staging.Events evn
join vaers_staging.Vaccines vac on evn.VAERS_ID = vac.VAERS_ID
where evn.DIED = true
order by evn.VAX_DATE desc
limit 12
```

### Full Outer Join to determine vaccines which result in adverse symptoms of blindness among patients

```python
%%bigquery
select VAX_NAME, VAX_TYPE, VAX_MANU
from vaers_staging.Vaccines vac 
full outer join vaers_staging.Symptoms sym on vac.VAERS_ID = sym.VAERS_ID
where SYMPTOM1 = 'Blindness'
order by vac.VAX_NAME 
limit 12
```
