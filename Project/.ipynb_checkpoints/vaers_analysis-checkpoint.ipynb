### VAERS Data Analysis

#### Show relationship between number of days spent in hospital and if patient died. As number of days in hospital increases, deaths decrease. View created for Google Data Studio accessibility.

```python
%%bigquery
create or replace view vaers_modeled.v_hosp_deaths as
select HOSPDAYS, count(DIED) as num_died
from `studied-brand-266702.vaers_modeled.Adverse_Event_Beam_DF`
group by HOSPDAYS
having HOSPDAYS <= 20
order by HOSPDAYS
```

```python
%%bigquery
select HOSPDAYS, count(DIED) as num_died
from vaers_modeled.Adverse_Event_Beam_DF
group by HOSPDAYS
having HOSPDAYS <= 20
order by HOSPDAYS
```

#### Join Vaccine table with Adverse_Events table (with Vaccination table as bridge) and look at how many deaths reported for each vaccine name. Top 20 shown for space.

```python
%%bigquery
select v.VAX_NAME, count(e.DIED) as num_died
from vaers_modeled.Vaccine v join vaers_modeled.Vaccination_Beam_DF vax
on v.VAX_ID = vax.VAX_ID
join vaers_modeled.Adverse_Event_Beam_DF e on vax.VAERS_ID = e.VAERS_ID
group by v.VAX_NAME
order by num_died desc
limit 20
```

#### Join Adverse_Events table with Vaccination table and see how many patients visited the ER after receiving vaccinations from a specific organization. Most ER visits stem from Unknown or Private facilities.

```python
%%bigquery
select vax.V_ADMINBY, count(e.ER_VISIT) num_er_visit
from vaers_modeled.Adverse_Event_Beam_DF e join vaers_modeled.Vaccination_Beam_DF vax
on e.VAERS_ID = vax.VAERS_ID
group by vax.V_ADMINBY
order by num_er_visit desc
```
